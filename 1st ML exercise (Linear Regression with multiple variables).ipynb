```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('hiring.csv')
df
```

```python
#function
def word_to_num(text):
    units = {
        'zero' : 0, 'one' : 1, 'two' : 2, 'three': 3, 'four': 4, 'five': 5,
        'six': 6, 'seven' : 7, 'eight' : 8, 'nine' : 9, 'ten' : 10,  'NaN': 0,
        'eleven' : 11, 'twelve' : 12, 'thirteen' : 13, 'fourteen': 14, 'fifteen': 15,
        'sixteen': 16, 'seventeen' : 17, 'eighteen' : 18, 'nineteen' : 19
    }
    tens = {
        'twenty' : 20, 'thirty': 30, 'forty': 40, 'fifty': 50,
        'sixty': 60, 'seventy' : 70, 'eighty' : 80, 'ninety' : 90,
    }
    text = text.lower().replace('-',' ').replace('and', '')
    words = text.split()
    
    total = 0
    current = 0
    for word in words:
        if word in units:
            current = current + units[word]
        elif word in tens:
            current = current + tens[word]
        elif word == 'hundred':
            current = current * 100
        elif word == 'thousand':
            current = current * 1000
            total = total + current
    

    return total + current

word_to_num('forty-five')
```

```python
def text2num(text):
    nums = {
        'zero' : 0, 'one' : 1, 'two' : 2, 'three': 3, 'four': 4, 'five': 5,
        'six': 6, 'seven' : 7, 'eight' : 8, 'nine' : 9, 'ten' : 10,  'NaN': 0,
        'eleven' : 11, 'twelve' : 12, 'thirteen' : 13, 'fourteen': 14, 'fifteen': 15,
        'sixteen': 16, 'seventeen' : 17, 'eighteen' : 18, 'nineteen' : 19
    }
    
    result = 0
    if text in nums:
        result = result + nums[text]
    return result
```

```python
converted_experience = df.experience.apply(text2num)
converted_experience
```

```python
df.experience = converted_experience 
df
```

```python
import math
test_median = math.floor(df.test.median())
test_median
```

```python
df.test = df.test.fillna(test_median)
df.test
```
